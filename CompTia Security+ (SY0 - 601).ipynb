# 02- Explaining Threat Actors and Threat Intelligence:

>> Vulnerability, Threat, and Risk:
   - Vulnerability:
     - Weakness that could be triggered accidentally or exploited interntionally to cause a security breach

   - Risk:
     - Is the Likelihood and Impact of a threat actor exploiting a vlun

# 05- Summarizing Basic Cryptographic Concepts:
 
> Cryptography is the art of making informaation secure by encoding it

> Cryptanalysis is the art of cracking Cryptographic systems

>> Cryptography  types:
   - Encryption 
   - Hashing
   - Digital certificates (PKI)


>> Hashing Algorithms:
   - Hashing is the simplest type of Cryptographic operation
   - Hash Algorithm is used to prove the Integrity of data
   - Used for Authentication or Integrity
   - It cann't be used to store or transmit data
   - The plaintext cann't be recovered from the Digest (Hashing is oneway)
   - Hashing algorithm produces a "Fixed Length" string from and "input Plaintext"
   - The output can be called (Checksum - Message Digest - Hash)
   - Hashing Algorithms are:
     - Secure Hash Algorithm (SHA1 - SHA2) "Strongest algorithm - 256 bit"
     - Message Digest Algorithm #5 (MD5) "Not safe - 128 bit"


>> Encryption Ciphers and Keys:
   - Encryption Algorithm is a type of cryptocraphic process that encodes data so that it can be recovered, or decrypted
   - It used in Storing and Transmitting data
   - Encryption is used to prove the Confidentiality of data
   - Symmetric Encryption "Using the same Key to encrypt and decrypt"
     - Symmetric Encryption is used for Confidentiality
     - Symmetric Encryption is referred as (single key - private key - shared secret)
     - It's very fast
     - Cann't be used for Authentication or Integrity

   - ASymmetric Encryption "Using 2 keys Public and Private keys"
     - Operating performed by 2 keys called "Key Pair"  
     - ASymmetric Encryption is used for Prove Identity, Confidentiality, and Authentication
     - You send the Public key to anyone to encrypt any data and so you only can decrypt this data by your Private key
     - Need high prossesing powers
     - Public key Algorithms (RSA "2048 bits"- ECC "256 bits")

# 06.1- Implementing Public Key Infrastructure (PKI):

>> Public key Cryptography: 
   - Is the ASymmetric Encryption
   - Solves the problem of Distributing Enctyption keys
   - To apply Confidentiality you should share the Public key to encrytpt the data, and your Private key to decrypt the data
   - To apply Authentication you should encrypt Your Signature with your Pirvate key and share it to be dectypted by the Public key
   

>> Public Key Infrastructure (PKI)
   - Aims to prove that owners of public keys are who they say they are
   - Anyone issuing public keys should obtain a Digital Certificate
   - The validity of the Certificate is guaranteed by a Certificate Authority (CA)
   - Digital Certificate is the Wrapper of the Public key

# 10.1- Impelementing Network Security Appliances:

>> Access Control Lists (ACLs):
   - A Packet filtering firewall is configured by specifying a group of rules, called "Access Control List (ACL)"
   - Each Rule defines a specific type of data packet and the appropriate action to take when the packet  matches the ruele
   - An Action can be either to Deny (Block or Drop the packet, and optionally log and event) or to Accept (Let the packet pass through the firewall)
   - A packet filterring firewall can inspect the headers of IP packets
   - The Rules can be based on the informationfound in the headers:
     - IP Filtering "Accept or Deny according to the SRC/DEST IP address"
     - Protocol ID/Type "TCP - UDP - ICMP - Routing protocols - ...."
     - Port Filtering/Security "Accept or Deny according to the SRC/DEST Port Number"


>> Stateless and Stateful Firewalls Operation:
   - Stateless Firewalls:
     - Basic Packet Filtering firewall
     - Doesn't preserve information about network sessions
     - Each packet is analyzed independently with no record of previously proccessed packets
     - This Type requires the least proccessing effort, but it can be vulnerable to attacks
     - Can interduce Problems in traffic flow, when some sort of load balancing is being used or when clients or servers need to use dynamically assigned ports
     - It's Slow but More Secure

   - Stateful Firewalls:
     - Tracking information about the Sessions
     - Blocking malicious attempts to start a bogus session
     - Session data is stored in a State Table
     - It's Fast but Less Secure
     - It occures at "Transport OR Application" layer
       - At the Transport layer "Examines the TCP 3 way Handshake"
       - The Firewall can be configured to respind to Flooding or Hijacking Attempts by Blocking SRC IP 
       
       - At the Application layer "Inspect Packets content"
       - Key feature is to verify the application protocol matches the port
       - Cannot examine encrypted data packets, unless configured with an SSL/TLS inspector


>> Firewall Appliances:
   - It's a Stand alone hardware firewall deployed to monitor passing traffic and can be deployed in 2 ways:

   - Routed (Layer 3)
     - Performs Forwarding between subnets
     - Each Interface connects to a different subnet to represent a zone
     - Work as a Router

   - Bridged (Layer 2)
     - Inspects traffic passing between 2 nodes (Router - Swithc)
     - Also called Transparent Mode
     - Work as a Switch


>> Virtual Firewalls:
   - Usually deployed within Data Centers and Cloud Services, in 3 ways:
  
   - Hypervisor Based:
     - Filtering functionality is built into the Hypervisor or cloud provisioning tool, you can use API to write the ACLs

   - Virtual Appliance:
     - Deploying a vendor firewall appliance instance using virtualization

   - Multiple Context:
     - Refers to multiple Virtual Firewall instances running on a hardware firewall appliance
     - Each context has a seperate interface and can perform a distinct filtering rule


> Intrusion Detection System (IDS) is a means of using software tools to provide real time analysis of either Network Traffic or System and Application Logs to detect suspicious traffic based on (Signature Based Detection - Behavioral Based Detection)

> Network Based IDS (NDIS) captures traffic via a Packet Sniffer, referred to as a Sesor, To identify Malicious Traffic and Displays alerts like (Snort - Suricata - Zeek/Bro)

> Intrusion Prevention System (IPS) is like IDS and can provice an Active Response to any Network Threats by (End TCP session - send TCP reset)

> Next Deneration Firewalls is the Old Firewall + IDS OR IPS

> Unified Threat Management (UTM) refers to a security product that centralizes many types of security controls (Firewall - Anti Malware - NIPS - Spam Filtering - Content Filtering - Data Loss Prevention - VPN - Cloud Access Gateway)

> Web Application Firewall (WAF) is designed to protect software running on web servers and their back end databases from code injection and DoS attacks (ModSecurity - NAXSI - Imperva - F5)

>> Security Information and Event Management (SIEM):
   - Core functionality is aggregate traffic data and Logs
   - It collects data by using Simple Network Management Protocol (SNMP)
   - Top SIEM Vendors (Splunk - LogRhythm - QRadar - .....)


> Security Orchestration, Automation, and Response (SOAR) is designed as a solution to the problem of the volume of alerts overwhelming analysts' ability to respond

# 19- Sumarizing Risk Management Concepts:

>> Risk Management Processes:
   - Identify Mission Essential Functions
   - Identify Vulnerability
   - Identify Threats
   - Analyze Business Impactss
   - Identify Risk Response


>> Risk Types:
   - Enternal (Natural Disasters, Personal Disasters, Accidents)
   - External (Contractors)
   - Multiparty 
   - Intellecutal Property (IP) Theft
   - Software Compliance / Licensing
   - Legacy Systems


>> Quantitative Risk Assessment:
   - Risk assessment aims to assgn concrete values to each risk factor
   - Single Loss Expectancy (SLE) "Amount lost in a single occurence risk"
   - Annualized Rate of Occurrence (ARO) "How many times of occurrence in a year"
   - Annualized Loss Expectancy (ALE) "Ammount lost over the year ALE = SLE * ARO"


>> Qualitative Risk Assessment:
   - Risk assessment avouds the complexity of the Quantitative approach and is focused on identifying significant risk factors risk factors
   - Seeks out people's opinions of which risk factors are significant
   - Assets categorized as "Irreplaceable - High - Medium - Low Value"
   - Risks categorized as "One off - Recurring - Critical - High - Medium - Low Probability"


>> Mission Essentil Functions:
   - Business Impact Analysis (BIA) 
   - Maximum Tolerable Downtime (MTD) "Working time without problems"
   - Recovery Time Objective (RTO) "Period after Disaster (Offline)"
   - Work Recovery Time (WRT) "Period to reintegrate the systems"


>> Single Points of Failure (SPoF):
   - Asset that caused the Entire workflow to fail if it's damaged
   - It can be mitigated by Provisioning Redundant Components


> Mean Time To Failure (MTTF) used for Non_Repairable Assets "Need to be replaced" and Equal to (Total Time / # Devices)

> Mean Time Between Failure (MTBF) Used for Repairable or Replaceable Assets and Equal to (Total Time / # Failures)

   - (tar -tf inventory.tar) to display the contents of the folder

   - (cat backuphash) to compare the 2 Hashes
