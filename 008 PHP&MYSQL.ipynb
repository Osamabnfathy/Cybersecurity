```python
<?php
  // Database Connection
  $host = "localhost";
  $user_name = "ghost";
  $pass = "1234";
  $DB_name = "users";
  $connection = mysqli_connect($host, $user_name, $pass, $DB_name);
  
  if(!$connection) {
    die("Connection Error: ".mysqli_connect_error());
  }
  
  // MYSQL built-in functions
  // COUNT function
  $query = "SELECT COUNT(`id`) AS cnt FROM `accounts`";
  $result = mysqli_query($connection, $query);
  $data = mysqli_fetch_assoc($result);
  echo "Count = ".$data['cnt']."<br>";

  // MAX function
  $query = "SELECT MAX(`id`) AS mx FROM `accounts`";
  $result = mysqli_query($connection, $query);
  $data = mysqli_fetch_assoc($result);
  echo "Max ID = ".$data['mx']."<br>";
  
  // MIN function
  $query = "SELECT MIN(`id`) AS mn FROM `accounts`";
  $result = mysqli_query($connection, $query);
  $data = mysqli_fetch_assoc($result);
  echo "Min ID = ".$data['mn']."<br>";
  
  // SUM function
  $query = "SELECT SUM(`id`) AS sm FROM `accounts`";
  $result = mysqli_query($connection, $query);
  $data = mysqli_fetch_assoc($result);
  echo "Sum of IDs = ".$data['sm']."<br>";
  
  // AVG function
  $query = "SELECT AVG(`id`) AS avr FROM `accounts`";
  $result = mysqli_query($connection, $query);
  $data = mysqli_fetch_assoc($result);
  echo "Average ID = ".$data['avr']."<br>"; 
  
  // UCASE function
  $query = "SELECT UCASE(`name`) AS uc FROM `accounts`";
  $result = mysqli_query($connection, $query);
  while($data = mysqli_fetch_assoc($result)){
    echo "Upper Case Name = ".$data['uc']."<br>";
  }
  
  // LCASE function
  $query = "SELECT LCASE(`name`) AS lc FROM `accounts`";
  $result = mysqli_query($connection, $query);
  while($data = mysqli_fetch_assoc($result)){
    echo "Lower Case Name = ".$data['lc']."<br>";
  }
  
  // MID function "display some characters from the name"
  $query = "SELECT MID(`name`, 1, 3) AS md FROM `accounts`";
  $result = mysqli_query($connection, $query);
  while($data = mysqli_fetch_assoc($result)){
    echo "First 3 letters of the Name = ".$data['md']."<br>";
  }
  
  mysqli_close($connection);
?>
```
