# Section 4: Operational Procedures (4.9)

>> Backup Types:
   - Full Backup 
     - Backup everything
     - High Backup Time
     - Low Restore Time

   - Differential Backup
     - Full Backup first
     - Subsequent backups contain data changed since the last full backup
     - Restoration Requirs the Full Backup and The last Differential backup
     - Moderate Backup Time
     - Moderate Restore Time

   - Incremental Backup:
     - Full Backup first
     - Subsequent backups contain data changed since the last full backup and last Incremental backup
     - Restoration Requirs the Full Backup and all the Incremental backups
     - Low Backup Time
     - High Restore Time

   - Synthetic Backup:
     - Compination of the Incremental Backups and the Full Backup to make a Synthetic Full Backup
     - Low Backup Time
     - Low Restore Time


>> Grandfather-Father-Sone (GFS) Backup Method:
   - Each Month Backup is a Grandfather (Full Backup)
   - Each Week Backup is a Father (Full Backup)
   - Each Day Backup is a Son (Differential or Incremental Backup)


>> 3-2-1 Backup rule:
   - 3 Copies of data should be available (one Primary and 2 Backups)
   - 2 Different types of media should be used (Local Drive, Tape Backup, NAS)
   - 1 Copu of the backup should be offsite (Offsite storage, Cloud Backup)
