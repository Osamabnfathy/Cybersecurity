# Section 1: Mobile Devices (1.4)

> Small Outline Dual In-Line Memory Module (SO-DIMM) used in laptop

>> Storages types:
   - Magnetic disk (HDD)
   - Solid-State drive (SSD)
   - M.2


>> Liquid Crystal Display (LCD):
   - Lightweight
   - Low power
   - Inexpensive
   - Black levels are a challenge
   - Requires separate backlight


>> Twisted Nematic LCD (TN):
   - Original LCD technology
   - Fast response times (gaming)
   - Poor viewing angles 
   - Color shifts


>> In Plane Switching LCD (IPS):
   - Excellent color representation
   - More Expensive to produce than TN


>> Vertical Alignment LCD (VA):
   - Come between IPS and TN
   - Good color representation
   - Slower response times than TN


>> Organic Light Emitting Diode (OLED):
   - Organic compound emits light
   - Thinner and Lighter
   - Flexible and mobile
   - No glass needed
   - Very accurate color representation
   - A bit higher cost than LCD


> Digitizer use a pen as input

>> Universal Serial Bus (USB-C):
   - 24-pin double-sided USB connector
   - Used for both hosts and devices
   - Acts as USB 2.0/3.0/3.1/4
   - Can transmit other signals "Display port - HDMI - Thunderbolt"


>> 2G Networks:
   - Global System for Mobile Comm. (GSM)
   - Code Division Multiple Access (CDMA)


>> 3G Networks:
   - Bandwidth improved
   - Add GPS
   - Add Mobile television
   - Add Video on demand


>> 4G and Long Term Evolution (LTE):
   - Converged standard (GSM and CDMA)
   - Based on GSM and EDGE (Enhanced Data Rates for GSM Evolution)
   - Download rates up to 150 Mbit/s
   - LTE Advanced (LTE A) 300 Mbit/s


>> 5G Networks:
   - Higher frequencies
   - Speeds 10 Gbit/s
   - Larger data transfer
   - But Bandwidth becomes less


>> Global Positioning System (GPS):
   - Precise navigation need at least 4 satellites
   - Determines location based on Longitude, Latitude, and Altitude
   - Know location based on GPs, WiFi and cellular towers

# Section 2: Networking (2.8)

>> Connectionless protocols (UDP)
   - Dynamic Host Configuration Protocol (DHCP)
   - Trivial File Transfer Protocol (TFTP)


>> Connection oriented protocols (TCP)
   - Hypertext Transfer Protocol Secure (HTTPS)
   - Secure Shell (SSH)

 
>> Ports of IP addresses:
   - port 80 = unencrypted web data (HTTP)
   - port 443 = encrypted web data (HTTPS)
   - port 123 = time server ()
   - port 25 = mail server (SMTP)


> TCP port Nums aren't the same as UDP port Nums

>> File Transfer Protocol (FTP):
   - TCP/20 = Active mode data 
   - TCP/21 = Control
   - Transfer files between systems


>> Secure Shell (SSH):
   - TCP/22 = Encrypted Communication link
   - Looks and acts the same as Telnet


>> Telecommunication Network (Telnet):
   - TCP/23
   - Login to devices remotely
   - Not best choise production systems


>> Simple Mail Transfer Protocol (SMTP):
   - TCP/25
   - Server to server email transfer
   - Device to server email transfer
   - Other protocols used for receive emails (IMAP - POP3)


>> Domain Name System (DNS):
   - UDP/53
   - Converts names to IP addresses


>> Dynamic Host Configuration Protocol (DHCP):
   - UDP/67
   - UDP/68
   - Automated configuration of IP address, subnet mask and other options


>> Hypertext Transfer Protocol (HTTP - HTTPS):
   - TCP/80 = HTTP 
   - TCP/443 = HTTPS (with encryption)
   - Communication in the browser


>> Post Office Protocol version 3 (POP3):
   - TCP/110
   - Receive emails from a email server
   - Basic mail transfer funcitonality


>> Internet Message Access Protocol V4 (IMAP4):
   - TCP/143
   - Receive emails from a email server
   - Includes management of email inbox from multiple clients


>> Server Message Block (SMB):
   - Protocl used by Microsoft Windows
   - file hsaring, printer sharing
   - Called Commen Internet File System (CIFS)
   - USing Network Basic Input/Output System over TCP/IP (NetBIOS)
   - UDP/137 = NetBOIS name services (nbname)
   - TCP/139 = NetBOIS session service (nbsession)
   - Direct SMB comm. over TCP without the NetBOIS transport (TCP/445 NetBOIS-less)


>> Simple Network Management Protocol (SNMP):
   - UDP/161 = Queries
   - UDP/162 = Traps
   - Gather statistics from network devices


>> Lightweight Directory Acces Protocol (LDAP):
   - TCP/389
   - Store and retrieve information in a network directory
   - Used in Microsoft Actice Directory


>> Remote Desktop Protocol (RDP):
   - TCP/3389
   - Share a desktop from remote location 


>> Routers:
   - Traffic between IP subnets
   - Routers inside of switches called (layer 3 switches)
   - Forwards traffic based on IP address


>> Switches:
   - Bridging done in hardware through Application-specific integrated Circuit (ASIC)
   - Forwards traffic based on Data link address (MAC address)


>> Managed Switching:
   - VLAN support
   - Voice Traffic gets a higher proirity
   - Redundancy support by Spanning Tree Protocol (STP)
   - Port mirroring
   - External management (SNMP)


>> Access point:
   - It's not a wirless router
   - It's a bridge that extends the wired network onto wireless network
   - Forwards traffic based on Data link address (MAC address)


>> Firewalls:
   - Filters traffic by IP address and Port number
   - Found in OSI layer 4 (TCP/UDP)
   - Can encrypt traffic into/out of network
   - Most Firewalls can be layer 3 devices (Routers)


>> Hub (Multi-port repeater):
   - Traffic going in one port is repeated to every other ports
   - Everything is half-duplex
   - Less efficient as traffic increases


>> Cable modem (TV cable):
   - Transmission across multiple frequencies
   - Different traffic types
   - Data Over Cable Service Interface Specification (DOCSIS)
   - High-speed networking
   - Multiple services (Data - Voice - Video)


>> Digital Subscriber Line (DSL):
   - Asymmetric DSL (ADSL)
   - Uses telephone lines
   - Download speed is faster than upload speed


>> Optical Network Terminal (ONT):
   - Fiber to the premises
   - Connect the ISP fiber network to the copper network


>> Software Defined Networking (SDN):
   - Networking devices have different functional planes of operation
   - Split functions into separate logical units 
   - Extend functionality and management of a single device
   - Perfectly built for the cloud
   
   - Infrastructure layer (Data plane)
     - Process the network frames and packets
     - Forwarding, trunking, encrypting, NAT
   
   - Control layer (Control plane)
     - Manges actions of the data plane
     - Routing tables, session  tables, NAT tables
     - Dynamic routing protocol updates
   
   - Application layer (Management plane)
     - Configure and manage devices
     - SSH, browser, API, SNMP


>> 802.11a
   - Uses a 5GHz
   - Speed 54 Mb/s per channel
   - Small range


>> 802.11b
   - Uses a 2.4GHz
   - Speed of 11 Mb/s per channel
   - better range than 802.11a


>> 802.11g
   - Upgrade to 802.11b
   - Uses 2.4GHz
   - Speed of 54 Mb/s per channel


>> 802.11n (Wi-Fi 4):
   - Update to 802.11(a - b - g) 
   - Uses 2.4GHz / 5GHz
   - 40 MHz channel bandwidth
   - Speed of 150 Mb/s per channel
   - Uses 4 Multiple-Input Multiple-Output (MIMO)


>> 802.11ac (Wi-Fi 5):
   - Improvments over 802.11n (Wi-Fi 4)
   - Uses 5GHz
   - 160 MHz channel bandwidth
   - Faster Data transfer
   - Speed of 867 Mb/s per channel
   - Uses 8 DL Multi user MIMO (MU-MIMO)


>> 802.11ax (Wi-Fi 6):
   - Successor to  802.11ac (Wi-Fi 5)
   - Uses 2.4GHz / 5GHz
   - 20, 40, 80, 160 MHz channel bandwidth
   - 1.201 Mb/s per channel
   - Uses 8 DL/UL Bi-Directional (MU-MIMO)
   - Orthogonal Frequency-Division Multiple Access (OFDMA) as phone comm.


>> Radio-Frequency Identification (RFID):

>> Near-field communication (NFC):

>> File server:

>> Print server:
   - Server Massage Block (SMB)
   - Internet Printing Protocol (IPP)
   - Line Printer Daemon (LPD)


>> Syslog: 
   - Standard for message logging
   - Integrated into the SIEM
   - need alot of disk space


>> All-in-one security appliance:
   - Next generation firewall
   - Called Unified Threat Management (UTM)
   - Web Security Gateway
   - URL filter / Content inspection
   - Malware inspection
   - Spam filter
   - CSU / DSU
   - Router, Switch

>> Load balancer features:
   - Configurable load (Distribution)

>> Proxy server:
   - An intermediate server
   - Client makes the request to the proxy, then the proxy perfoms the actual request, finally the proxy provides results back to the client
   - Access control, Caching, URL filtering, content scanning


>> Supervisory Control and Data Acquisition System (SCADA): 
>> Multi-site Industrail Control Systems (ICS):
   - PC manages equipment

>> Automatic Private IP Addressing (APIPA):
   - No forwarding by routers
   - Reserved IPs (169.254.0.0 - 169.254.255.255)
   - first and last 256 IPs are reserved so (169.254.0.1 - 169.254.254.255)
   - NO DHCP server here
   - Uses the ARP to confirm the address isn't currently in use
   - You can't communicate outside this network


>> Address records (A / AAAA):
   - Defines the IP address of a host
   - A records are for IP v4 addresses
   - AAAA records are for IP v6

>> Sender Policy Framework (SPF):
   - A list of all servers authorized to send emails for this domain
   - Prevent mail spoofing
   - Mail servers perform a check to see if incoming mail did come from an authorized host


>> Virtual Local Area Networks (VLANs):
   - Agroup of devices in the same broadcast domain
   - Seperated logically instead of physically
   - Allows multiple networks on one physical link


>> Virtual Private Networks (VPNs):
   - Encrypted data traversing a public network
   - Often integrated into a firewall


>> Cable broadband (TV cable):
   - Transmission across multiple Freq.
   - Different traffic types
   - Data Over Cable Service Interface Specification (DOCSIS)
   - High speed Networking
   - Multiple services (Data - Vioce - Video)


>> Storage Area network (SAN):
   - Looks like local storage device
   - Block level access
   - very efficient reading and writing
   - May use an isolated network and high-speed network tecknologies


>> Tone Generator:
   - Toner Probe:
     - Wher does that wire go?
     - Follow the tone

   - Tone Generator:
     - Connect the tone generator to the wire
     - Puts an analog sound on the wire

   - Inductive Probe:
     - Used to locate the sound
     - Doosn't need to touch the copper
     - Hear through a small speaker
     - Easy wire tacing


>> Cable Tester:
   - Continuity test
   - simple wire map
   - can identify missing pins or crossed wires


>> Loopback plugs:
   - Useful for testing physical ports
   - Send and Receive data in same device to see if they match or not
   - Done in "Ethernet - T1 - Fiber"
   - Loop done in pairs of (1,3 - 2,6 - 4,7 - 5,8)


>> Taps and port mirrors:
   - Send a copy to a packet capture device
   - Put in the middle between source and destination
   - Passive tap for Fiber
   - Active tap for Copper
   - Port Mirror 
     - Port redirection, Switched Port ANalyzer (SPAN)

   - DB-15 connector

   - Vidoe and audio
   - Digital signal

   - Digital information sent in packetized form
   - It's like Ethernet and PCI Express

>> Flash drives:
   - Electrically Erasable Programmable Read Only Memory (EEPROM)
   - Non volatile memory
   - No power required to retain data 
   - Limited number of writes and still read the data only
   - Not designed for archival storage 
   - Easy to lose or damage


>> Redundant Array of Independent Disks (RAID):
   - RAID 0 (Striping):
     - 2 Drives or more 
     - We split data between them
     - Data written quickly
     - Failure of one break the system

   - RAID 1 (Mirroring):
     - Two identical disks are used
     - Every file is duplicated
     - Failure not affect the system

   - RAID 5 (Striping with Parity):
     - Same as RAID 0 with Parity block 
     - Parity block can rebuild data 
     - Dat is avialable after Failure

   - RAID 10 (Stripe of Mirrors):
     - It uses RAID 0 and RAID 1 together
     - we need atleast 4 Drives


> Unified Extensible Firmware Interface (UEFI BOIS). Based on Intel's Extensible Firmware Interface (EFI), Designed to replace the legacy BIOS
